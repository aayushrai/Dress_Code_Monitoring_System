```python
import tkinter as tk
import cv2
import PIL
import PIL.Image, PIL.ImageTk
import numpy as np
from tkinter.ttk import *
from tkinter import *
from tkinter.filedialog import askdirectory
from tkinter import messagebox
import os
from PIL import Image
from imageai.Detection import ObjectDetection
import os
import cv2
import numpy as np
from keras.models import load_model
```

```python
#object detection
execution_path = os.getcwd()
detector = ObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath( os.path.join(execution_path , "yolo.h5"))
detector.loadModel()
```

```python
#Dress Code Classifier
model = load_model("/home/uchiha/Desktop/Dress_code/Dress_code_classifier_model/models/version8.h5")
```

```python
class Runner(tk.Tk):


    def __init__(self, *args, **kwargs):
        tk.Tk.__init__(self, *args, **kwargs)
        container = tk.Frame(self)

        container.pack(side="top", fill="both", expand=True)
        self.frames = {}

        frame = StartPage(container, self)

        self.frames[StartPage] = frame
        frame.grid(row=0, column=0, sticky="nsew")
        self.show_frame(StartPage)

    def show_frame(self, cont):
        frame = self.frames[cont]
        frame.tkraise()
        
class StartPage(tk.Frame):

    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.folder_path = None
        self.images_names_list = []
        self.image_index = -1
        self.img_width = 1150
        self.img_height =950
        self.threshold = 0.7
        self.crop = None
        
        #frame 1
        buttons_frame = tk.Frame(self)
        self.btn_open=Button(buttons_frame, text="Open\nFolder",width=10,command = lambda:self.open_file())
        self.btn_open.pack(anchor=tk.CENTER,padx =30,pady =30)
        self.btn_next=Button(buttons_frame, text="Next\nImage", width=10,command = lambda:self.next_image())
        self.btn_next.pack(anchor=tk.CENTER,padx =30,pady =30)
        self.btn_prev=Button(buttons_frame, text="Previous\nImage", width=10,command = lambda:self.prev_image())
        self.btn_prev.pack(anchor=tk.CENTER,padx =30,pady =30)
        buttons_frame.pack(pady=30,side="left",anchor="nw")
        #frame 2
        cap_frame2 = tk.Frame(self)
        self.canvas1 = tk.Canvas(cap_frame2, width = self.img_width, height =self.img_height,bg="black",bd=2)
        self.canvas1.pack(padx=5,pady=5)
        cap_frame2.pack(side="left",anchor="nw",pady=30)
        
        #frame 3
        cap_frame3 = tk.Frame(self)
        self.cropCanvas = tk.Canvas(cap_frame3, width = 170, height =300,bg="black",bd=2)
        self.cropCanvas.pack(padx=5,pady=5)

        cropDetectBtn = tk.Frame(cap_frame3)
        self.crop=Button(cropDetectBtn, text="CROP", width=10,state=DISABLED,command = lambda:self.cropImage())
        self.crop.pack(side="left",anchor=tk.CENTER)
        self.detector=Button(cropDetectBtn, text="DETECT", width=10,state=DISABLED,command = lambda:self.detect())
        self.detector.pack(anchor=tk.CENTER)
        cropDetectBtn.pack()
        
        cropDetectBtn = tk.Frame(cap_frame3)
        forml = tk.Frame(cropDetectBtn)
        formalLabel = tk.Label(forml, text="Formal  ")
        formalLabel.pack(side="left")
        self.formalBar = ttk.Progressbar(forml, style='text.Horizontal.TProgressbar', length=300,maximum=100, value=0)
        self.formalBar.pack(anchor="nw",pady=20,padx=10,side="left")
        self.formalProb = tk.Label(forml, text="")
        self.formalProb.pack(side="left")
        forml.pack()
        informl = tk.Frame(cropDetectBtn)
        informalLabel = tk.Label(informl, text="Informal")
        informalLabel.pack(side="left")
        self.informalBar = ttk.Progressbar(informl, style='text.Horizontal.TProgressbar', length=300,maximum=100, value=0)
        self.informalBar.pack(anchor="nw",pady=20,padx=10,side="left")
        self.informalProb = tk.Label(informl, text="")
        self.informalProb.pack(side="left")
        informl.pack(side="left")
        cropDetectBtn.pack(pady=21)
        
        self.console_text = tk.Text(
            cap_frame3, fg="green", bg="black", width=55, height=26, state=tk.DISABLED)
        scrollbar = tk.Scrollbar(cap_frame3, command=self.console_text.yview)
        scrollbar.pack(side="right", fill=tk.Y)
        self.console_text.pack(expand=1, fill=tk.BOTH)
        self.console_text['yscrollcommand'] = scrollbar.set
        cap_frame3.pack(side="left",anchor="nw",pady=30,padx=20)
        self.output_to_console("Choose folder....")
        
    def output_to_console(self, new_text):
        self.console_text.config(state=tk.NORMAL)
        self.console_text.insert(tk.END,"\n")
        self.console_text.insert(tk.END, ">> " + new_text + "\n")
        self.console_text.see(tk.END)
        self.console_text.config(state=tk.DISABLED)
        
    def changeThreshold(self):
        self.threshold = self.threshold_var.get()
        self.detect()

    def open_file(self): 
        self.folder_path =askdirectory()
        self.image_index = -1
        self.images_names_list = []
        if self.folder_path:
            for name in os.listdir(self.folder_path):
                if name.split(".")[-1].lower() in ["jpg","jpeg","png"]:
                    self.images_names_list.append(os.path.join(self.folder_path,name))
            if len(self.images_names_list) == 0:
                messagebox.showinfo("Warning","No image in selected folder, please select folder again.")
            else:
                self.output_to_console("Click On Crop....")
            self.next_image()
        self.crop["state"] ="normal"
        
            
    def next_image(self):
        if self.image_index < len(self.images_names_list)-1:
            self.image_index += 1
            self.img = cv2.imread(self.images_names_list[self.image_index])
            self.img = cv2.cvtColor(self.img,cv2.COLOR_BGR2RGB)
            self.img = cv2.resize(self.img,(self.img_width,self.img_height))
            self.canvas_img = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(self.img))
            self.canvas1.create_image(self.img_width//2,self.img_height//2, image = self.canvas_img, anchor = tk.CENTER)

    def prev_image(self):
        if self.image_index >0:
            self.image_index -= 1
            self.img = cv2.imread(self.images_names_list[self.image_index])
            self.img = cv2.cvtColor(self.img,cv2.COLOR_BGR2RGB)
            self.img = cv2.resize(self.img,(self.img_width,self.img_height))
            self.canvas_img = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(self.img))
            self.canvas1.create_image(self.img_width//2, self.img_height//2, image = self.canvas_img, anchor = tk.CENTER)
    
    def cropImage(self):
        self.output_to_console("Cropping started wait its take some time....")
        slot1 = cv2.imread(self.images_names_list[self.image_index])
        detections = detector.detectObjectsFromImage(
        input_image=self.images_names_list[self.image_index],
        output_image_path=os.path.join(execution_path, "out.jpg"))
        print(detections)
        person =False
        if detections:
            for eachObject in detections:
                print(eachObject["name"], " : ", eachObject["percentage_probability"])
                print(eachObject)
                if eachObject["name"] == "person":
                    box = eachObject["box_points"]
                    for k in range(len(box)):
                        if box[k] < 0:
                            box[k] = 0
                    img = slot1[box[1]:box[3], box[0]:box[2]]
                    person = True
        if person:
            self.crop = img
            resized_image = cv2.resize(img,(170+5,300+5))
            cvt = cv2.cvtColor(resized_image,cv2.COLOR_BGR2RGB)
            self.cropResized = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(cvt))
            self.cropCanvas.create_image((170//2),(300//2), image = self.cropResized)
            self.detector["state"] ="normal"
        else:
            messagebox.showinfo("Warning","No person detected in image, please choose another image")
        cvt = cv2.cvtColor(slot1,cv2.COLOR_BGR2RGB)
        cvt = cv2.resize(cvt,(self.img_width,self.img_height))
        self.canvas_img = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(cvt))
        self.canvas1.create_image(self.img_width//2, self.img_height//2, image = self.canvas_img, anchor = tk.CENTER)   
        self.output_to_console("Click On Detect....")

    def detect(self):
        # the array based representation of the image will be used later in order to prepare the
        # result image with boxes and labels on it.
        # Actual detection.
        h = 300
        w = 150
        img = cv2.resize(self.crop,(w,h))
        img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        img.shape
        pred = model.predict(img.reshape(1,h,w,1))
        clas = model.predict_classes(img.reshape(1,h,w,1))
        class_label = None
        if clas[0] == 0:
            class_label = "FORMAL"
        elif clas[0] == 1:
            class_label = "INFORMAL"
        self.formalBar['value']=pred[0][0]*100
        self.informalBar['value']=pred[0][1]*100
        self.formalProb.config(text="[{:.2f}%]".format(pred[0][0]))
        self.informalProb.config(text="[{:.2f}%]".format(pred[0][1]))
        self.output_to_console("Prediction:"+str(class_label))


        
app = Runner()
app.mainloop()
```
